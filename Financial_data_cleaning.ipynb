```python

import pandas as pd
import numpy as np
df = pd.read_csv('Financial Analytics data.csv')
df
```

```python
df.drop_duplicates(subset=['S.No.'], keep='first', inplace=True)

```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Data Cleaning
# Drop unnecessary columns
df.drop(columns=['S.No.', 'Unnamed: 4'], inplace=True)

df.fillna(0.0, inplace=True)

# Descriptive Analysis
print(df.describe())
```

```python

# Step 4: Convert 'Sales Qtr - Crore' to string and then standardize
df['Sales Qtr - Crore'] = df['Sales Qtr - Crore'].astype(str).str.replace(',', '').str.replace('crore', '').astype(float)

```

```python
# Step 5: Format Market Cap and Sales Columns as Currencies
df['Market Capitalization'] = df['Mar Cap - Crore'].map('{:,.2f} Cr'.format)
df['Quarterly Sale'] = df['Sales Qtr - Crore'].map('{:,.2f} Cr'.format)

# Step 6: Handle Missing Values
df['Quarterly Sale'].fillna(0, inplace=True)  # Replace NaN values with 0 for sales
```

```python

df.drop(columns=['Mar Cap - Crore', 'Sales Qtr - Crore'], inplace=True)
df
```

```python
df['Market Capitalization'] = df['Market Capitalization'].str.replace(' Cr', '').str.replace(',', '').astype(float)
```

```python
# prompt: change dtype tofloat 'Quarterly Sale'

df['Quarterly Sale'] = df['Quarterly Sale'].str.replace(' Cr', '').str.replace(',', '').astype(float)
```

```python
df
# Save cleaned data to a new CSV file
df.to_csv('cleaned_data.csv', index=False)
```
