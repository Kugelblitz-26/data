```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score

# Define a function to train and evaluate a regression model
def train_and_evaluate_model(model, X_train, y_train, X_test, y_test):
    # Train the model
    model.fit(X_train, y_train)

    # Make predictions
    y_pred = model.predict(X_test)

    # Calculate mean squared error
    mse = mean_squared_error(y_test, y_pred)
    print(f"Mean Squared Error: {mse:.2f}")

    # Calculate R-squared score
    r2 = r2_score(y_test, y_pred)
    print(f"R-squared Score: {r2:.2f}")

    # Perform cross-validation and calculate the mean score
    cv_score = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    cv_rmse = np.sqrt(-cv_score).mean()
    print(f"Cross-Validation RMSE: {cv_rmse:.2f}")
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso

# Linear Regression
linear_regression = LinearRegression()
print("Linear Regression:")
train_and_evaluate_model(linear_regression, X_train, y_train, X_test, y_test)
```

```python
# Ridge Regression
ridge_regression = Ridge()
print("Ridge Regression:")
train_and_evaluate_model(ridge_regression, X_train, y_train, X_test, y_test)
```

```python
# Lasso Regression
lasso_regression = Lasso()
print("Lasso Regression:")
train_and_evaluate_model(lasso_regression, X_train, y_train, X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
# Decision Tree Regressor
decision_tree = DecisionTreeRegressor()
print("Decision Tree Regressor:")
train_and_evaluate_model(decision_tree, X_train, y_train, X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
# Random Forest Regressor
random_forest = RandomForestRegressor()
print("Random Forest Regressor:")
train_and_evaluate_model(random_forest, X_train, y_train, X_test, y_test)

# Extra Trees Regressor
extra_trees = ExtraTreesRegressor()
print("Extra Trees Regressor:")
train_and_evaluate_model(extra_trees, X_train, y_train, X_test, y_test)
```

```python

models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "Decision Tree Regression": DecisionTreeRegressor(),
    "Random Forest Regression": RandomForestRegressor(),
    "Extra Trees Regression": ExtraTreesRegressor(),
    "LGBM Regressor": LGBMRegressor(),
    "XGBoost Regressor": XGBRegressor()
}


def train_and_evaluate(model, X_train, y_train, X_test, y_test):
    # Train the model
    model.fit(X_train, y_train)

    # Make predictions on the test data
    y_pred = model.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    return mse, r2

# Dictionary to store model results
model_results = {}


for model_name, model in models.items():
    print(f"Training and evaluating {model_name}...")

    # Train and evaluate the model
    mse, r2 = train_and_evaluate(model, X_train, y_train, X_test, y_test)

    model_results[model_name] = {"MSE": mse, "R2 Score": r2}

# Display the results
print("\nModel Evaluation Results:")
for model_name, results in model_results.items():
    print(f"{model_name}:")
    print(f"Mean Squared Error (MSE): {results['MSE']:.4f}")
    print(f"R2 Score: {results['R2 Score']:.4f}")
    print("-" * 50)
```
