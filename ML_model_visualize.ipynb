# ML MODELS VISUALIZE

LINEAR REGRESSION

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Number of bedrooms
y = np.array([120000, 150000, 170000, 220000, 250000])  # House prices

# Create and train a linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the data points
plt.scatter(X, y, color='blue', label='Actual Prices')

# Plot the linear regression line
plt.plot(X, y_pred, color='red', label='Linear Regression')

# Labels and legend
plt.xlabel('Number of Bedrooms')
plt.ylabel('House Price')
plt.legend()

# Show the plot
plt.show()
```

lOGISTIC REGERESSION


```python

from sklearn.linear_model import LogisticRegression

# Sample data
X = np.array([[2, 3], [4, 5], [5, 6], [7, 8]])  # Features (e.g., two features per data point)
y = np.array([0, 0, 1, 1])  # Classes (0 or 1)

# Create and train a logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Decision boundary
x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.01), np.arange(x2_min, x2_max, 0.01))
Z = model.predict(np.c_[xx1.ravel(), xx2.ravel()])
Z = Z.reshape(xx1.shape)

# Plot the decision boundary
plt.contourf(xx1, xx2, Z, alpha=0.4, cmap='RdYlBu')

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='RdYlBu')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Show the plot
plt.show()
```

DECISSION TREE

```python

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create and train a decision tree classifier
model = DecisionTreeClassifier()
model.fit(X, y)

# Visualize the decision tree
plt.figure(figsize=(12, 6))
plot_tree(model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

RENDOM FOREST CLASSIFIER

```python

from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create and train a random forest classifier
model = RandomForestClassifier()
model.fit(X, y)

# Get feature importances
importances = model.feature_importances_
feature_names = iris.feature_names

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.barh(range(len(importances)), importances, tick_label=feature_names)
plt.xlabel('Feature Importance')
plt.title('Feature Importance in Random Forest')
plt.show()
```

SUPPORT VECTOR MACHINE

```python

from sklearn.datasets import make_classification
from sklearn.svm import SVC

# Generate synthetic data for binary classification
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Create and train an SVM classifier
model = SVC(kernel='linear', C=1.0)
model.fit(X, y)

# Plot the decision boundary and support vectors
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid to evaluate model
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM Decision Boundary and Support Vectors')
plt.show()
```

K-MEANS CLUSTERING

```python

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data for clustering
X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Create and train a K-Means clustering model
model = KMeans(n_clusters=4)
model.fit(X)

# Visualize cluster centers and data points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=model.labels_, cmap='viridis')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()
```

PRINCIPLE COMPONENT ANALYSIS
DIMENSIONALITY REDUCTION

```python

from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create a PCA model with 2 components
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Plot the reduced data
plt.figure(figsize=(8, 6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Dimensionality Reduction')
plt.show()
```

RANDOM FOREST DECISION BOUNDARIES

```python

from sklearn.datasets import make_moons
from sklearn.ensemble import RandomForestClassifier

# Generate synthetic data for binary classification
X, y = make_moons(n_samples=100, noise=0.3, random_state=42)

# Create and train a Random Forest classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Create a mesh grid to plot decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundaries
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap='viridis', alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Random Forest Decision Boundaries')
plt.show()

```

```python

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data for clustering
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Create and train a K-Means clustering model with different numbers of clusters
k_values = [2, 3, 4]
plt.figure(figsize=(15, 5))

for i, k in enumerate(k_values):
    model = KMeans(n_clusters=k)
    model.fit(X)

    # Plot the clusters and cluster centers
    plt.subplot(1, len(k_values), i + 1)
    plt.scatter(X[:, 0], X[:, 1], c=model.labels_, cmap='viridis')
    plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=300, c='red', marker='x')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(f'K-Means with {k} Clusters')

plt.show()
```

HIERARCHICAL CLUSTERING

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import linkage, dendrogram

# Generate synthetic data for clustering
X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Perform hierarchical clustering
Z = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(8, 6))
dendrogram(Z, truncate_mode='lastp', p=4)
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

NEURAL NETWORK

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
import requests
from PIL import Image
import io  # Import the 'io' module

# Load the VGG16 model
model = VGG16(weights='imagenet')

# Define a function to classify and visualize an image
def classify_image(url):
    response = requests.get(url)
    img = Image.open(io.BytesIO(response.content))
    img = img.resize((224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    preds = model.predict(x)
    label = decode_predictions(preds, top=1)[0][0]

    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.title(f'Predicted: {label[1]} ({label[2]:.2f})')
    plt.axis('off')
    plt.show()

# Example usage:
image_url = 'https://images.pexels.com/photos/1490908/pexels-photo-1490908.jpeg?'
classify_image(image_url)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create a PCA model with 3 components
pca = PCA(n_components=3)
X_reduced = pca.fit_transform(X)

# Plot the reduced data in 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y, cmap='viridis')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('PCA Dimensionality Reduction in 3D')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

# Generate synthetic data for binary classification in 3D
X, y = make_classification(n_samples=100, n_features=3, n_informative=3, n_redundant=0, random_state=42)

# Create and train a Logistic Regression classifier
model = LogisticRegression()
model.fit(X, y)

# Create a mesh grid to visualize decision boundaries in 3D
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
z_min, z_max = X[:, 2].min() - 1, X[:, 2].max() + 1

xx, yy, zz = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01), np.arange(z_min, z_max, 0.01))
XYZ = np.c_[xx.ravel(), yy.ravel(), zz.ravel()]
Z = model.predict(XYZ)
Z = Z.reshape(xx.shape)

# Plot the decision boundaries and data points in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.contourf(xx, yy, zz, Z, cmap='viridis', alpha=0.4)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('Logistic Regression Decision Boundaries in 3D')
plt.show()
```
